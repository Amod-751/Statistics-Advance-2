```python
'''
Question 1: What is hypothesis testing in statistics?
    ->Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It helps us determine whether there is enough evidence to support or reject a specific assumption (called a hypothesis) about the population.
    Key Components of Hypothesis Testing

       => Null Hypothesis (H₀):

        The default assumption or claim we want to test.

        Example: "There is no difference in average sales before and after the new marketing campaign."

        => Alternative Hypothesis (H₁ or Ha):

        What we want to prove or show evidence for.

        Example: "There is a difference in average sales before and after the new marketing campaign."

       => Test Statistic:

        A value calculated from sample data (e.g., t-test, z-test, chi-square test) that measures how far the sample result is from what we expect under the null hypothesis.

        Significance Level (α):

        A threshold (commonly 0.05 or 5%) that defines the probability of rejecting the null hypothesis when it is actually true (Type I error).

       => p-value:

        The probability of obtaining results as extreme as (or more extreme than) the observed results, assuming the null hypothesis is true.

        If p-value ≤ α, reject H₀.

        If p-value > α, fail to reject H₀.
        
        
    Steps in Hypothesis Testing

       => State the hypotheses

        H₀: No effect/difference.

        H₁: Effect/difference exists.

       => Choose significance level (α)

        Commonly 0.05.

       => Select appropriate test

        z-test, t-test, chi-square, ANOVA, etc.

       => Compute test statistic & p-value

       => Make a decision

        If p ≤ α → reject H₀ (evidence supports H₁).

        If p > α → fail to reject H₀ (not enough evidence against H₀).
        
        
     Example

        Suppose a company claims the average weight of its product is 500g.

        H₀: μ = 500g (claim is true)

        H₁: μ ≠ 500g (claim is false)

        Collect sample → calculate mean and standard deviation → perform t-test.

        If p-value < 0.05, reject H₀ → evidence suggests the product weight is not 500g.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
    ->1. Null Hypothesis (H₀):

        The default assumption or claim about a population parameter.

        It states that there is no effect, no difference, or no relationship in the data.

        It represents the status quo or "nothing new is happening."

         Example:

        A company claims the average weight of a product is 500g.

        H₀: μ = 500g

     2. Alternative Hypothesis (H₁ or Ha):

        The statement we want to test or find evidence for.

        It suggests that there is an effect, a difference, or a relationship.

        If data strongly contradicts H₀, we support H₁.

         Example:

        Continuing with the company’s product:

        H₁: μ ≠ 500g (the average weight is different from 500g).
        
       
     3. Key Differences:  
       
        | Aspect                              | Null Hypothesis (H₀)                      | Alternative Hypothesis (H₁/Ha)                            
| ----------------------------------- | ----------------------------------------- | --------------------------------------------------------- |
| Meaning                        | No effect, no difference, status quo      | There is an effect, a difference, or relationship         |
| Goal                            | Assume true until evidence says otherwise | Try to prove or find evidence for                         |
| Example (Product weight = 500g) | μ = 500g                                  | μ ≠ 500g (two-tailed) or μ > 500g / μ < 500g (one-tailed) |
| Decision                        | Fail to reject or reject (never "accept") | Supported only if H₀ is rejected                          |


    4. Types of Alternative Hypotheses

        Two-tailed test: H₁: μ ≠ μ₀ (any difference, higher or lower).

        Right-tailed test: H₁: μ > μ₀ (greater than).

        Left-tailed test: H₁: μ < μ₀ (less than).
        
        
        
Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
    ->The significance level (α) is the probability of making a Type I error — rejecting the null hypothesis (H₀) when it is actually true.

        It is a threshold set before conducting the test that determines how much evidence we require to reject H₀.

        Common choices: 0.05 (5%), 0.01 (1%), 0.10 (10%).

      Role in Hypothesis Testing:

        => Compare p-value with α

        If p ≤ α → reject H₀ (evidence supports H₁).

        If p > α → fail to reject H₀ (not enough evidence against H₀).

        => Controls the risk of false positives

        A smaller α → stricter test, less chance of wrongly rejecting H₀.

        A larger α → more lenient test, higher chance of rejecting H₀.

    Example

        Suppose a pharmaceutical company claims that a new drug has no effect on blood pressure.

        H₀: The drug has no effect.

        H₁: The drug reduces blood pressure.

        Set α = 0.05.

        After testing, we get p = 0.03.

        Since p (0.03) ≤ α (0.05) → Reject H₀.

        Conclusion: The drug likely reduces blood pressure.

        If instead p = 0.07, then p > α, so we fail to reject H₀ (not enough evidence to prove the drug works).


Question 4: What are Type I and Type II errors? Give examples of each.
    ->1. Type I Error (False Positive)

        Definition: Rejecting the null hypothesis when it is actually true.

        Probability of this error = α (significance level).

         Examples:

        A medical test says a healthy person has a disease.

        A court convicts an innocent person.

        A company concludes its new drug works when in fact it does not.

     2. Type II Error (False Negative)

        Definition: Failing to reject the null hypothesis when it is actually false.

        Probability of this error = β.

        Power of a test = 1 – β (ability to detect a true effect).

         Examples:

        A medical test fails to detect a disease in a sick person.

        A court acquits a guilty person.

        A company concludes its new drug does not work when in fact it does.
        
   =>Analogy (Courtroom Example):

    H₀: The defendant is innocent.
    H₁: The defendant is guilty.
    
        | Decision | Reality: Innocent                         | Reality: Guilty                           
    | -------- | ----------------------------------------- | ----------------------------------------- |
    | Convict  | ❌ Type I Error (innocent person punished) | ✅ Correct                                 |
    | Acquit   | ✅ Correct                                 | ❌ Type II Error (guilty person goes free) |



Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
    ->Both are statistical tests used to compare sample data against a population or between samples, but they differ mainly in assumptions about sample size and population standard deviation.
     1. Z-test:
        A hypothesis test used when the population variance (σ²) or standard deviation (σ) is known, and the sample size is large (n ≥ 30).

     Key Points:

        Based on standard normal distribution (Z-distribution).

        Assumes data follows a normal distribution (or approximately normal for large n due to Central Limit Theorem).

        More accurate when σ is known.

     The formula for a one-sample Z-test is:

        Z = (X̄ − μ₀) / (σ / √n)

     Use Cases:

        Testing whether the average height of students differs from a known population mean when population σ is known.

        Comparing proportions in large samples.

     2. T-test
        A hypothesis test used when the population variance (σ²) is unknown and the sample size is small (n < 30).

     Key Points:

        Based on Student’s t-distribution, which has heavier tails than the Z-distribution (to account for extra uncertainty in small samples).

        As n increases, t-distribution approaches the normal distribution (so for very large n, t ≈ z).

        The formula for a one-sample t-test is:

        t = (X̄ - μ₀) / (s / √n)


        where s = sample standard deviation.

     Types of T-tests:

        One-sample t-test → compare sample mean to population mean.

        Independent two-sample t-test → compare means of two independent groups.

        Paired t-test → compare means of two related samples (before vs after).

     Use Cases:

        Testing whether a small class of 20 students has an average score different from 50 (population σ unknown).

        Comparing average blood pressure before and after treatment in the same patients (paired t-test).


'''
```

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Step 1: Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Step 2: Hypothesized population mean
mu_0 = 50

# Step 3: Compute Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Step 4: Compute p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 5: Print results
print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Sample Size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Step 6: Decision at α = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H₀: Evidence suggests the mean is different from 50.")
else:
    print("Fail to reject H₀: No strong evidence that the mean differs from 50.")

```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # true mean and standard deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Step 3: Calculate 95% confidence interval
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # t critical value
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print("Sample Mean:", sample_mean)
print("95% Confidence Interval: ({:.3f}, {:.3f})".format(ci_lower, ci_upper))

# Step 5: Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='dashed', linewidth=2, label=f"Upper 95% CI = {ci_upper:.2f}")
plt.title("Normal Distribution Data with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_zscores(data):
    """
    Function to calculate Z-scores for a dataset and plot histogram of standardized data.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    
    # Calculate Z-scores
    z_scores = [(x - mean) / std for x in data]
    
    # Print summary
    print("Mean of data:", mean)
    print("Standard Deviation of data:", std)
    print("First 5 Z-scores:", z_scores[:5])
    
    # Plot histogram of Z-scores
    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7, density=True)
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.legend()
    plt.show()
    
    return z_scores

# Example dataset
data = [10, 12, 15, 18, 20, 22, 25, 28, 30, 35]
z_scores = calculate_zscores(data)
```
